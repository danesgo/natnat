```python
#Battery State of Charge based on consumption
initial_SoC_peten = 60
initial_SoC_coban = 60
OnePercentEnergy = 271.0125/100 #301.125Wh for a 30 cell battery pack
Battery_Year_status_peten=[initial_SoC_peten]
Battery_Year_status_coban=[initial_SoC_coban]
hours_off_peten = 0
hours_off_coban = 0
hours_overcharge_peten = 0
hours_overcharge_coban = 0
contador_peten = 0
contador_coban = 0
maximum_peten = 0
maximum_coban=0
accumulated_hours_off_peten = []
accumulated_hours_off_coban = []
year = [0]

for x in range(1,len(Real_year_power_obtained_peten)):
    var1=Battery_Year_status_peten[x-1]+(Real_year_power_obtained_peten[x]*0.5-LED_consumption[int(x%48)]-USB_consumption[int(x%48)]-TV_consumption[int(x%48)])/OnePercentEnergy
    var2=Battery_Year_status_coban[x-1]+(Real_year_power_obtained_coban[x]*0.5-LED_consumption[int(x%48)]-USB_consumption[int(x%48)]-TV_consumption[int(x%48)])/OnePercentEnergy
    Battery_Year_status_peten.append(var1)
    Battery_Year_status_coban.append(var2)
    year.append(x)
    if Battery_Year_status_peten[x]>90:
        Battery_Year_status_peten[x]=90
        hours_overcharge_peten = hours_overcharge_peten +1
        contador_peten=0
    elif Battery_Year_status_peten[x]<10:
        Battery_Year_status_peten[x]=10
        contador_peten = contador_peten+1
        hours_off_peten = hours_off_peten+1
        if contador_peten/2>=maximum_peten:
            maximum_peten=contador_peten/2
    elif Battery_Year_status_peten[x]>10 and contador_peten>0:
        accumulated_hours_off_peten.append(contador_peten/2)
        contador_peten=0
        
        
    if Battery_Year_status_coban[x]>90:
        Battery_Year_status_coban[x]=90
        hours_overcharge_coban = hours_overcharge_coban +1
        contador_coban = 0
    elif Battery_Year_status_coban[x]<10:
        Battery_Year_status_coban[x]=10
        hours_off_coban = hours_off_coban+1
        contador_coban = contador_coban+1
        if contador_coban/2>=maximum_coban:
            maximum_coban = contador_coban/2
    elif Battery_Year_status_coban[x]>10 and contador_coban>0:
        accumulated_hours_off_coban.append(contador_coban/2)
        contador_coban=0
        
average_time_off_peten = 0
for x in xrange(len(accumulated_hours_off_peten)):
    average_time_off_peten = average_time_off_peten+accumulated_hours_off_peten[x]

average_time_off_peten = average_time_off_peten/len(accumulated_hours_off_peten)

average_time_off_coban = 0
for x in xrange(len(accumulated_hours_off_coban)):
    average_time_off_coban = average_time_off_coban+accumulated_hours_off_coban[x]

average_time_off_coban = average_time_off_coban/len(accumulated_hours_off_coban)

print "Horas que pasa la refri apagada en Peten: ",hours_off_peten/2
print "Horas que pasa la refri apagada en Coban: ", hours_off_coban/2
print "Maximo de horas en Peten: ",maximum_peten
print "Maximo de horas en Coban: ",maximum_coban
print "Lapsos de tiempo de apagado en Peten: ",average_time_off_peten
print "Lapsos de tiempo de apagado en Coban: ",average_time_off_coban
print "Cantidad de veces que se apaga el sistema en Peten: ",len(accumulated_hours_off_peten)
print "Cantidad de veces que se apaga el sistema en Coban: ",len(accumulated_hours_off_coban)
plt.rcParams["figure.figsize"]=[30,20]
fig, ax = plt.subplots(nrows = 2, ncols=1)
fig.subplots_adjust(hspace = 0.3)    #Definir espacios
major_ticks = np.arange(0,100,10)
minor_ticks = np.arange(0,100,5)
ax[0].set_yticks(major_ticks)
ax[0].set_yticks(minor_ticks,minor=True)
ax[1].set_yticks(major_ticks)
ax[1].set_yticks(minor_ticks,minor=True)
ax[0].grid(which='major',alpha=0.8)
ax[0].grid(which='minor',alpha = 0.5)
ax[1].grid(which='major',alpha=0.8)
ax[1].grid(which='minor',alpha = 0.5)
ax[0].set_title("Irradiance Peten")
ax[1].set_title("Irradiance Coban")
ax[0].plot(year,Battery_Year_status_peten)
ax[1].plot(year,Battery_Year_status_coban)
```
